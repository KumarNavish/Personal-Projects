```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 1)
```

```python
x1 = [1,2,3,4,5,6]
x2 = [7.5,11,16,18,20,26]
x = np.c_[x1,x2]
x
```

```python
plt.scatter(x1,x2)
plt.show()
```

```python
x_transformed = pca.fit_transform(x)
```

```python
x_transformed
```

```python
pca.components_
```

```python
x_approx = pca.inverse_transform(x_transformed)
```

```python
plt.scatter(x_approx[:,0],x_approx[:,1])
plt.show()
```

```python
np.random.seed(0)
```

```python
mean1 = np.array([0,0,0])
cov1 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class1 = np.random.multivariate_normal(mean1,cov1,100)
```

```python
mean2 = np.array([1,1,1])
cov2 = np.array([[1,0,0],[0,1,0],[0,0,1]])
class2 = np.random.multivariate_normal(mean2,cov2,100)
```

```python
from mpl_toolkits.mplot3d import Axes3D,proj3d

fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection = '3d')
ax.plot(class1[:,0],class1[:,1],class1[:,2],'o')
ax.plot(class2[:,0],class2[:,1],class2[:,2],'^')
```

```python
pca = PCA(n_components = 2)
```

```python
x_comb = np.concatenate((class1,class2))
x_comb[:2]
```

```python
trans_data = pca.fit_transform(x_comb)
trans_data[0:6]
```

```python
plt.scatter(trans_data[:,0],trans_data[:,1])
plt.show()
```

```python
pca.components_
```

```python
np.concatenate(([[1,2,3],[2,5,9]],[[0,5,0],[4,9,9]]))
```

```python
plt.scatter(trans_data[:100,0],trans_data[:100,1])
plt.scatter(trans_data[100:,0],trans_data[100:,1])
plt.show()
```

```python
x_approx = pca.inverse_transform(trans_data)
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111,projection = '3d')
ax.plot(x_approx[:100,0],x_approx[:100,1],x_approx[:100,2],'o')
ax.plot(x_approx[100:,0],x_approx[100:,1],x_approx[100:,2],'go')
```
